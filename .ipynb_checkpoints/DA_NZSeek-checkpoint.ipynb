```python
def deduplication(x):
    trim = x.strip()
    index = (trim + trim).find(trim,1)
    if index != -1:
        return trim[:index]
    else:
        return trim
```

```python
admin['Location'] = admin['Location'].apply(deduplication)
admin['Location']
```

```python
admin['Area'].fillna(value = 'Unknown', inplace = True)
```

```python
def apply_posted_time(x):
    if 'd' in x:
        return -int(x[:-1])
    if 'm' in x:
        return -(int(x[:-1])*30)
    else:
        return 0
```

```python
admin['Post Time'] = admin['Post Time'].apply(apply_posted_time)
```
