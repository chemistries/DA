# Data Visualization

## Top 5 suburbs median price

```python
df_price.groupby('Suburb').agg({'Price':'median'}).sort_values(by = 'Price', ascending = False).iloc[:5].plot(kind = 'bar')
```

## Number of properties by types

```python
df_price.groupby('Type')['Suburb'].agg('count').plot(kind = 'pie')
```

## Top 5 Sellers 

```python
df_price.groupby('SellerG')['Suburb'].agg('count').sort_values(ascending = False).iloc[:5].plot(kind = 'bar')
```

## Top 5 Listed suburbs

```python
df_price.groupby('Suburb')['Suburb'].agg('count').sort_values(ascending = False).iloc[:5].plot(kind = 'bar')
```

## Different methods proportion 

```python
df_price.groupby('Method')['Suburb'].agg('count').plot(kind = 'pie')
```
